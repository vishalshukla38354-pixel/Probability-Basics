Question 1: A die is rolled. What is the probability of getting:

(a) An even number

(b) A number greater than 4

**Answer** -
      
         A die has 6  outcomes:{1,2,3,4,5,6}
     
     (a) Probability of getting an even number

     Even number ={2,4,6} ➡ 3 outcomes
     P(even) = 3/6=1/2

     (b) Probability of getting a number greater than 4

     Numbers greater than 4 ={5,6} ➡ 2

     P (numbee > 4) = 2/6 =1/3

Question 6: The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what
is the standard error of the mean (SEM)?

**Answer** -

Population standard deviation σ = ₹5,000

Sample size n = 100

Standard Error of Mean

SEM = σ/ √n

SEM + 5000/√100 = 5000/10 = 500

Answer  = ₹500
